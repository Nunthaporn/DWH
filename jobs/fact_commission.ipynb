```python
def calculate_condi_com_status(row):
    if pd.notna(row["numpay"]) and row["numpay"] < 2:
        return "หารเท่าทุกงวด"
    
    if pd.notna(row["condition_install"]) and str(row["condition_install"]).startswith("0000"):
        return "หารเท่าทุกงวด"
    
    if (
        pd.notna(row["numpay"]) and row["numpay"] == 10 and
        pd.notna(row["money_one"]) and pd.notna(row["money_ten"]) and
        row["money_one"] == row["money_ten"]
    ):
        return "หารเท่าทุกงวด"
    
    if row.get("no_car") == "730":
        if pd.notna(row["condition_install"]) and str(row["condition_install"]).strip() != "":
            prefix = str(row["condition_install"])[:4]
            return f"งวดแรกจ่าย {prefix}"
        else:
            return "หารเท่าทุกงวด"

    return None
```

```python
df_merged["condi_com_status"] = df_merged.apply(calculate_condi_com_status, axis=1)
```

```python
import pandas as pd

# สร้างคอลัมน์ condi_com ตามเงื่อนไข
def calculate_condi_com(row):
    try:
        if (
            row["percent_install"] == 0
            or (
                row["percent_install"] == 0 and row["isGovernmentOfficer"] == 1 and pd.notna(row["chanel_main"]) and "ผ่อน" in row["chanel_main"]
            )
            or (
                row["current_campaign"] == "ผ่อนไม่หักคอม" and pd.notna(row["id_cus"]) and not row["id_cus"].startswith("FTR")
            )
        ):
            return "ไม่หักคอม"

        if pd.notna(row["id_cus"]) and row["id_cus"].startswith("FTR"):
            return "ปกติ"

        if pd.notna(row["chanel_main"]) and "ผ่อน" not in row["chanel_main"]:
            return "ปกติ"

        elif (
            pd.notna(row["datestart"])
            and row["datestart"] > pd.to_datetime("2024-07-17")
            and row["status_vip"] in ("yes", "yes-new")
            and row["numpay"] > 1
            and row["condition_install"] == "000003"
            and str(row["percent_install"]) in ["0.3", "0.5", "0.9", "1.2", "1.5", "1.8", "2.1", "2.4", "3", "1", "2", "2.5", "3.5", "4", "4.5", "5"]
        ):
            return "หักคอม 0.3,0.5"
        else:
            return "หักคอม"
    except Exception:
        return "หักคอม"

# สร้างคอลัมน์ใหม่
df_merged["condi_com"] = df_merged.apply(calculate_condi_com, axis=1)
```

```python
import pandas as pd

# แปลงให้เป็นตัวเลขก่อน
df_merged["numpay"] = pd.to_numeric(df_merged["numpay"], errors="coerce")
df_merged["discom"] = pd.to_numeric(df_merged["discom"], errors="coerce")
df_merged["show_com_ins"] = pd.to_numeric(df_merged["show_com_ins"], errors="coerce")

# สร้างคอลัมน์ commission ตามเงื่อนไข
df_merged["commission"] = df_merged.apply(
    lambda row: row["discom"] if row["numpay"] > 1 else row["show_com_ins"],
    axis=1
)

# แสดงตัวอย่างผลลัพธ์
print(df_merged[["numpay", "discom", "show_com_ins", "commission"]].head())
```

```python
import pandas as pd
import numpy as np

# แปลงช่องว่างทุกแบบเป็น NaN ชั่วคราว เพื่อการนับข้อมูล
df_temp = df.replace(r'^\s*$', np.nan, regex=True)

# เพิ่มคอลัมน์ช่วยนับจำนวนข้อมูล (non-null)
df['non_empty_count'] = df_temp.notnull().sum(axis=1)

# >>>> ส่วนที่แก้ไขตรงนี้ <<<<
# ตรวจสอบ agent_id ที่ไม่ว่าง (ไม่ใช่ NaN และไม่ใช่ช่องว่าง)
valid_agent_id_mask = df['quotation_num'].astype(str).str.strip().ne('') & df['quotation_num'].notna()

# แยกกลุ่มที่ agent_id ไม่ว่างและ agent_id ว่าง
df_with_id = df[valid_agent_id_mask]
df_without_id = df[~valid_agent_id_mask]

# คัดแถวที่ agent_id ซ้ำ โดยเก็บแถวที่มีข้อมูลมากที่สุด
df_with_id_cleaned = df_with_id.sort_values('non_empty_count', ascending=False).drop_duplicates(subset='quotation_num', keep='first')

# รวมกลับ
df_cleaned = pd.concat([df_with_id_cleaned, df_without_id], ignore_index=True)

# ลบคอลัมน์ช่วย
df_cleaned = df_cleaned.drop(columns=['non_empty_count'])
df_cleaned = df_cleaned.replace(
    to_replace=r'^\s*$|(?i:^none$)|^-$',  # << แก้ตรงนี้
    value=np.nan,
    regex=True
)


df_cleaned.columns = df_cleaned.columns.str.lower()
df_cleaned
```
